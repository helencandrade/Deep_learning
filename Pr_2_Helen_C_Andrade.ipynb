```python
from numpy import genfromtxt
cluster_data = genfromtxt('cluster_data.csv', delimiter=',')
cluster_label = genfromtxt('cluster_label.csv', delimiter=',')
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(cluster_data[:, 0], cluster_data[:, 1], cluster_data[:, 2], marker="o", c= cluster_label, s=25, edgecolor="r")
plt.show()
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(cluster_data, cluster_label)
log_reg = LogisticRegression(random_state=50)
log_reg.fit(x_train, y_train)
```

```python
print("coef dados treinamento {}".format(log_reg.coef_))
```

```python
from sklearn.metrics import accuracy_score
pred = log_reg.predict(x_test)
pred_train = log_reg.predict(x_train)
score_treinamento = accuracy_score(pred_train,y_train)
score_teste = accuracy_score(pred,y_test)
print(" Acuracia dados treinamento :",score_treinamento)
print(" Acuracia dados teste :",score_teste)
```

SVM

```python
from sklearn.svm import SVC
```

```python
clf = SVC(kernel="linear", C=float("inf"))
clf.fit(x_train, y_train)
print("coef:", clf.coef_)
```

```python
svc_pred_train = clf.predict(x_train)
svc_pred_test = clf.predict(x_test)
print("Acuracia dados treinamento:", accuracy_score(svc_pred_train, y_train))
print("Acuracia dados de teste:", accuracy_score(svc_pred_test, y_test))
```
